```python
import numpy as np
```

```python
def findnth(haystack, needle, n):
    parts= haystack.split(needle, n+1)
    if len(parts)<=n+1:
        return -1
    return len(haystack)-len(parts[-1])-len(needle)
```

```python
for question in range(1, 10):
    fname = "q" + str(question) + ".txt"
    with open("./DataStudyEvaluation-Qs/" + fname) as f:
        content = f.readlines()
    # you may also want to remove whitespace characters like `\n` at the end of each line
    content = [x.strip() for x in content] 

    for participant in range(len(content)):
        line = content[participant]
        count = 0
        while (findnth(line, "[", count) > -1):
            start = findnth(line, "[", count)
            end = findnth(line, "]", count)
            count = count + 1
            
            answer = line[start+1:end]
            answer = answer.replace("\t", "    ")
            print(answer, "- P" + str(participant+1), "/ Q" + str(question))
```
