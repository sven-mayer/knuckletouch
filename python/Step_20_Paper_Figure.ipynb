```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df = pd.read_pickle("DataStudyCollection/df_statistics.pkl")
```

```python
for y in df.Input.unique():
    for x in df.userID.unique():
        img = df[(df.userID == x) & (df.Input == y)].iloc[100].Blobs[1:7, 1:7]
        plt.axis('off')
        plt.imshow(img)
        plt.savefig("./out/blob_%s_%i.pdf" % (y,x) , bbox_inches='tight', transparent=False, pad_inches=0)
        plt.show()
```
